```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
tasks = ['LinearSmoothPursuit', 'ArcSmoothPursuit', 'RapidVisualSearch', 'RapidVisualSearchAvoidance', 'Mean']
model_names = ['MLP', 'LSTM', 'Vector', 'Vector-P']
scores_mlp = [0.00002835, 0.00002451, 0.00011005, 0.00010000, 0.00006573]
scores_lstm = [0.00008646, 0.00007347, 0.00043981, 0.00039520, 0.00024873]
scores_vector = [0.00280687, 0.00369038, 0.00562071, 0.00544502, 0.00439074]
scores_vector_p = [0.00002428, 0.00002386, 0.00012286, 0.00012458, 0.00007389]
model_scores = [scores_mlp, scores_lstm, scores_vector, scores_vector_p]
```

```python
gap = 0.1
w = (1 - gap) / (len(tasks) + 1)
colors = [f'C{i:d}' for i in range(len(tasks))]

xs = [[start + (gap / 2) + (w * (i + 1)) for i in range(len(tasks))]
      for start in range(len(model_scores))]

plt.figure(figsize=(10.8, 7.2))
for i, scores in enumerate(model_scores):
    plt.bar(xs[i], scores, w - 0.02, color=colors, label=tasks if i == 0 else None)
plt.title('Test Loss per Task')
plt.ylabel('Test Loss (MSE)')
plt.xlabel('Model')
plt.xticks([0.5 + i for i in range(len(model_scores))], model_names)
plt.legend()
plt.show()
```
