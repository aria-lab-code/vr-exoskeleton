# train

```python
import numpy as np

from vr_exoskeleton import data_utils, train
```

```python
data_utils.TASK_NAMES
```

```python
suite_codes = ['linear', 'arc', 'search', 'avoid', 'all']
suites = [[task] for task in data_utils.TASK_NAMES] + [data_utils.TASK_NAMES]
suites
```

```python
seeds = tuple(range(40, 45))
seeds
```

## MLP

```python
seed_suite_losses_test = list()  # (n, 5, 5)
for seed in seeds:
    suite_losses_test = list()  # (5, 5)
    for code_train, task_names_train in zip(suite_codes, suites):
        losses_test, tasks_test = train.train(
            'mlp',
            run_name=f'90hz-angle-h8-8-train_{code_train}',
            seed=seed,
            ignore_users={'User21'},
            task_names_train=task_names_train,
            hidden_sizes=[8, 8],
            lr=0.0001,
        )  # (5)
        suite_losses_test.append(losses_test)
    seed_suite_losses_test.append(suite_losses_test)

seed_suite_losses_test = np.array(seed_suite_losses_test)
np.mean(seed_suite_losses_test, axis=0)
```

```python
seed_suite_losses_test_lstm = list()  # (n, 5, 5)
for seed in seeds:
    suite_losses_test = list()  # (5, 5)
    for code_train, task_names_train in zip(suite_codes, suites):
        losses_test, tasks_test = train.train(
            'lstm',
            run_name=f'45hz-angle-h16-train_{code_train}',
            seed=seed,
            ignore_users={'User21'},
            task_names_train=task_names_train,
            downsampling_rate=2,
            hidden_sizes=[16],
            lr=0.0001,
        )  # (5)
        suite_losses_test.append(losses_test)
    seed_suite_losses_test_lstm.append(suite_losses_test)

seed_suite_losses_test_lstm = np.array(seed_suite_losses_test_lstm)
np.mean(seed_suite_losses_test_lstm, axis=0)
```
